### 2. Using Wordcloud for Visualisation

```python
#Positive Sentiments
posSentiment = dataframe[ dataframe['sentiment'] == 'Positive']
posSentiment = posSentiment['text']
posSentiment.head() #to see the first few rows of the dataset
```

```python
def WordCloudVisualise(somedata, color = 'lightblue'):
    initial_words = ' '.join(somedata)
    temp_words = initial_words.split()
    final_words = []
    for i in temp_words:
        if((not i.startswith('@')) and (not i.startswith('#')) and ('http' not in i) and (i != 'RT')) :
            final_words.append(i)
    final_words = " ".join(final_words)
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate( final_words)
    plt.figure(1,figsize=(16, 16))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
WordCloudVisualise(posSentiment)
```
